```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
y = pd.read_csv("drive/MyDrive/AICTE dataset/mental-and-substance-use-as-share-of-disease.csv")
y.head()
```

```python
df = pd.read_csv("drive/MyDrive/AICTE dataset/mental-and-substance-use-as-share-of-disease.csv")
df = df.iloc[:, :3]
df.head()
```

```python
y = y.iloc[:, 3]
```

```python
y
```

```python
x = pd.read_csv("drive/MyDrive/AICTE dataset/prevalence-by-mental-and-substance-use-disorder.csv")
x.head()
```

```python
x = x.iloc[:, 3:]
x.head()
```

```python
data = pd.concat((x, y), axis='columns')
data.head()
```

```python
data.set_axis(["Scizophrenia", "Bipolar Disorder", "Eating Disorder", "Anxiety", "Drug Usage", "Depression", "Alcohol", "Mental_Fitness"], axis='columns', inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(), annot=True, cmap='Blues')
plt.plot()
```

```python
sns.pairplot(data, corner=True)
plt.show()
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
x = pd.concat((df, x), axis='columns')
x.head()
```

```python
x = x.drop(["Code"], axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_train_fin, x_test_fin = x_train.iloc[:, 2:], x_test.iloc[:, 2:]
```

```python
x_train_fin
```

```python
from sklearn.svm import SVR
svr = SVR(kernel = 'rbf')
svr.fit(x_train_fin, y_train)
svr.score(x_test_fin, y_test)
```

```python
y_pred = svr.predict(x_test_fin)
from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_pred, y_test))
```

```python
from sklearn.model_selection import GridSearchCV

# defining parameter range
param_grid = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf']}

grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3)

# fitting the model for grid search
grid.fit(x_train_fin, y_train)
```

```python

# print best parameter after tuning
print(grid.best_params_)

# print how our model looks after hyper-parameter tuning
print(grid.best_estimator_)
```

```python
svr = SVR(C=1000, gamma=1, kernel='rbf')
svr.fit(x_train_fin, y_train)
svr.score(x_test_fin, y_test)
```

```python
y_pred = svr.predict(x_test_fin)
from sklearn.metrics import mean_squared_error
from math import sqrt
sqrt(mean_squared_error(y_pred, y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion='poisson',random_state=0)
regressor.fit(x_train_fin, y_train)
regressor.score(x_test_fin, y_test)
```

```python
y_pred = regressor.predict(x_test_fin)
sqrt(mean_squared_error(y_pred, y_test))
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_fin, y_train)
model.score(x_test_fin, y_test)
```

```python
y_pred = model.predict(x_test_fin)
sqrt(mean_squared_error(y_pred, y_test))
```
